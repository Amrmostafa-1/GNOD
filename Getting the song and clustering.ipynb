#### Extract all the items

```python
def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify", playlist_id)
    tracks = results['items']
    count = 0  # Initialize count before the loop
    while results['next'] is not None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep_time = randint(1, 3000) / 1000  # Calculate sleep time
        sleep(sleep_time)  # respectful nap
        count += 1  # Increment count
        print(f"{count}, slept for {sleep_time} seconds")
    return tracks
```

```python
all_tracks = get_playlist_tracks("6sx8rpqIE3R9sfGSb8Udxy")
len(all_tracks)
```

#### Flatten the data

```python
tracks1 = json_normalize(all_tracks)
```

```python
tracks1.head()
```

#### make the list of dictionary just a dictionary to be easier to deal with

```python
def list_to_dict(x):
    return {i: x[i] for i in range(len(x))}

tracks1['artist_dict'] = tracks1['track.artists'].apply(list_to_dict)
tracks1.head()
```

```python
tracks1['artist_dict'][0]
```

#### Convert this dictionary into a df which is inside a new column in the main df

```python
def expand_list_dict(row):
    df = pd.DataFrame.from_dict(row['artist_dict'], orient='index')
    df['song_id'] = row['track.id']
    return df

tracks1['artists_dfs'] = tracks1.apply(expand_list_dict, axis=1)
tracks1['artists_dfs'][0]
```

#### Extract it into a new df

```python
artist_df = pd.DataFrame(columns=['external_urls', 'href', 'id', 'name', 'type', 'uri'])
for mini_df in tracks1['artists_dfs']:
    artist_df = pd.concat([artist_df, mini_df], axis=0)
    
artist_df
```

#### Merge both df

```python
df_merged1 = pd.merge(left=tracks1,
                    right=artist_df,
                    how='inner',
                    left_on='track.id',
                    right_on='song_id')
df_merged1.head()
```

#### Choose only the columns we're intrested in

```python
df_final1 = df_merged1[['track.name', 'name', 'song_id']]
df_final1
```

### Get features of the songs

#### Get the features in chunks of 100 as spotify has a limit of 100 per request

```python
chunks = [(i, i+100) for i in range(0, len(df_final1), 100)]
audio_features_list = []
count = 0
for chunk in chunks:
    id_list100 = df_final1['song_id'][chunk[0]:chunk[1]]
    audio_features_list = audio_features_list + sp.audio_features(id_list100)
    sleep_time = (randint(1,3000)/1000)
    sleep(sleep_time)  # respectful nap
    count += 1  # Increment count
    print(f"{count}, slept for {sleep_time} seconds")
len(audio_features_list)
```

#### Flatten the feautres into a df

```python
audio_features_df = json_normalize(audio_features_list)
```
